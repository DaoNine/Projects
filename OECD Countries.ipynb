The dataframes in the current state are not useful.

Therefore, we have to restructure the datasets.

```python
wages = wages.pivot(index='TIME', columns='LOCATION',values='Value' )
wages.tail()
```

```python
whours=whours.pivot(index='TIME', columns='LOCATION', values='Value')
whours.tail()
```

### Let's have a look at the average working hours in the year 2016.

```python
#Create Series of the working hours in the year 2016
whours16 = pd.Series(data=whours.tail(1).values.tolist()[0], index=whours.columns.values.tolist()).sort_values()
whours16.dropna(inplace=True)
```

```python
#Create bar chart
color=np.array(['rgb(253,174,107)']*whours16.shape[0])
color[whours16.median()==whours16.values]='rgb(44,127,184)'
color[whours16.index=='DEU']='rgb(44,162,95)'
color[whours16.max()==whours16.values] ='rgb(44,127,184)'
color[whours16.index=='LUX'] = 'rgb(44,127,184)'

data = [go.Bar(
            x=whours16.index,
            y=whours16.values,
            marker=dict(
                color=color.tolist())
)]

layout = go.Layout(
    title='Average working hours of the OECD countries [2016]',
    xaxis = dict(title='OECD countries'), 
    yaxis = dict(title='yearly average working hours per person')
)

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

In the year 2016, German workers had the lowest average working hours of all OECD countries.

Whereas, Mexican workers had the highest average workings hours.

In general, working hours doesn't indicate the wealth of a country.

Income is a more suitable measurement for prosperity.

### What about the yearly average income?

```python
#Create series of the average wages in the year 2016
inc16 = pd.Series(data=wages.tail(1).values.tolist()[0], index=wages.columns.values.tolist()).sort_values()
```

```python
#Create the bar chart
color=np.array(['rgb(253,174,107)']*inc16.shape[0])
color[inc16.median()==inc16.values]='rgb(44,127,184)'
color[inc16.index=='DEU']='rgb(44,162,95)'
color[inc16.min()==inc16.values] ='rgb(44,127,184)'
color[inc16.max()==inc16.values] ='rgb(44,127,184)'

data = [go.Bar(
            x=inc16.index,
            y=inc16.values,
            marker=dict(
                color=color.tolist())
    )]

layout = go.Layout(
    title='Average wages of the OECD countries [2016]',
    xaxis = dict(title='OECD countries'), 
    yaxis = dict(title='Yearly average wages per person [$US]')
)

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

We saw in the chart before that Mexican workers had the longest average working hours, here we can see that they earn the lowest yearly average income of all OECD countries.

The yearly average wage in Luxembourg was the highest of all OECD countries, even though their average working hours were one of at the lower end.

In conclusion, productivity is the decisive factor of wealth.

### In which part of Europe was the income the highest in the year 2016?

```python
weuwages = wages[['LUX', 'NLD', 'AUT', 'CHE', 'BEL', 'DEU', 'FRA']]

eeuwages = wages[['POL', 'SVN', 'CZE', 'HUN']]

europe = pd.concat([eeuwages.mean(axis=1), neuwages.mean(axis=1), seuwages.mean(axis=1), weuwages.mean(axis=1)], axis=1)

europe

europe.rename({0:'East-Europe', 1:'North-Europe', 2:'South-Europe', 3:'West-Europe'}, axis=1, inplace=True)
europe.head()


trace0 = go.Scatter(
    x = europe.index,
    y = europe['East-Europe'],
    mode = 'lines',
    name = 'East-Europe'
)
trace1 = go.Scatter(
    x = europe.index,
    y = europe['North-Europe'],
    mode = 'lines',
    name = 'North-Europe'
)
trace2 = go.Scatter(
    x = europe.index,
    y = europe['South-Europe'],
    mode = 'lines',
    name = 'South-Europe'
)

trace3 = go.Scatter(
    x = europe.index,
    y = europe['West-Europe'],
    mode = 'lines',
    name = 'West-Europe'
)

data = [trace0, trace1, trace2, trace3]

layout = go.Layout(title='Average incomes of the average incomes', 
                  xaxis = dict(title='Year'), 
                  yaxis = dict(title='Yearly average income [$US]')
                  )

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

In average, countries in west Europe have higher salaries than in all other parts of europe.

The lowest average incomes are in East-European countries.
